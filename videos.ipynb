```python
import pandas as pd
file_path = r'D:/pythonpro/Python Videos for Python Pro Channel (Morning) - Pytho Pro.csv'
df = pd.read_csv(file_path)
df['Date (Published)'] = pd.to_datetime(df['Date (Published)'], dayfirst=True, format='%d %b %Y', errors='coerce')
problematic_rows = df[pd.isna(df['Date (Published)'])]
print("Problematic rows after date conversion:")
print(problematic_rows)
daily_upload_count = df.groupby(['Creator', df['Date (Published)'].dt.date]).size()
print("\nDaily Uploads by Creator:")
print(daily_upload_count)
```

# Bar chart (Daily uploads by creator)

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
daily_upload_count.unstack(level=0).plot(kind='bar', stacked=True)
plt.title('Daily Uploads by Creator')
plt.xlabel('Date')
plt.ylabel('Number of Uploads')
plt.xticks(rotation=45)
plt.legend(title='Creator')
plt.tight_layout()
plt.show()
```

# Pie chart (Daily uploads by creator)

```python
import matplotlib.pyplot as plt
daily_upload_count = df.groupby(['Creator', df['Date (Published)'].dt.date]).size()
plt.figure(figsize=(8, 8)) 
colors = ['gold', 'lightcoral', 'lightskyblue']  
daily_upload_count.groupby('Creator').sum().plot.pie(autopct='%1.1f%%', colors=colors)
plt.title('Daily Uploads Distribution by Creator')
plt.ylabel('')  
plt.legend(title='Creator', loc='upper right')
plt.tight_layout()
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
file_path = r'D:/pythonpro/Python Videos for Python Pro Channel (Morning) - Pytho Pro.csv'
df = pd.read_csv(file_path)
df['Date (Published)'] = pd.to_datetime(df['Date (Published)'], dayfirst=True, errors='coerce')
df = df.dropna(subset=['Date (Published)'])
df['Month'] = df['Date (Published)'].dt.month
df['Year'] = df['Date (Published)'].dt.year
monthly_upload_count = df.groupby(['Creator', 'Year', 'Month']).size().unstack(level=0)
monthly_upload_count.plot(kind='bar', figsize=(10, 6))
plt.title('Monthly Uploads by Creator')
plt.xlabel('Month')
plt.ylabel('Upload Count')
plt.xticks(rotation=0)
plt.legend(title='Creator')
plt.tight_layout()
plt.show()
```
